```python
import requests
import json
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        #elif type(x) is list:
        #    i = 0
        #    for a in x:
        #        flatten(a, name + str(i) + '_')
        #       i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
url_base = "http://0.0.0.0:8000/api/v1/characteristica_elastic/?format=json&final=true"
response = requests.get(url_base)
num_pages = response.json()["last_page"]
data = []
for page in range(2,num_pages +1):
    url_current = url_base + f"&page={page}"
    response = requests.get(url_current)
    data += response.json()["data"]["data"]

flatten_data = [flatten_json(d) for d in data]
df = pd.DataFrame(flatten_data)
```

```python
def results(x):
    records = x.drop(columns=['category', 'final',"ctype"]).to_dict("list") 
    return records
```

```python
indivduals = df.pivot_table(columns=["category"], index=["individual_pk","individual_name"], aggfunc=results)
groups = df.pivot_table(columns=["category"], index=["group_pk","group_name"], aggfunc=results)
```

```python
indivduals = indivduals.reset_index()
groups = groups.reset_index()
```

```python
df_group = pd.DataFrame([flatten_json(row) for row in groups.to_dict("records")])
```

```python
df_individual = pd.DataFrame([flatten_json(row) for row in indivduals.to_dict("records")])
```

```python
df = df_group.append(df_individual)
```

```python
final = df.dropna(axis='columns',how='all')
final.head()
```

```python
final.to_csv("test", sep="\t" )
```
